## Librerías y parámetros

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000 # time grid length
M = 80000 # number of simulations
t0 = 0 # initial time
T = 10 # final time
r = 0.01 # interest rate
sigma = 0.02 # volatility
S0 = 100 # initial price
K = 80 # Strike price
```

```python
t = np.linspace(t0, T, N) # time grid
deltaT = t[1] - t[0] 
```

## Simulación de las curvas


```python
# we compute the exponential using vector operations
exponential = np.exp((r-sigma**2/2) *deltaT + sigma*np.sqrt(deltaT)*np.random.randn(N-1,M)) 
# first column = initial conditions
S = np.vstack([S0 * np.ones((1,M)), exponential])
# We perform the cumulative product
S = np.cumprod(S,axis=0)
# Plot 30 curves to test
plt.plot(S[:,0:30]);
```

## Valoración de un call europeo

```python
euCall = lambda S: max(S[-1] - K,0)
prices = np.fromiter(map(euCall, S.T), dtype=np.float64)
plt.hist(prices, density=True, bins=int(np.sqrt(M)))
mu = np.mean(prices); sigma = np.std(prices)
print(f"mean={mu}, std={sigma}")
xaxis = np.linspace(mu-3*sigma, mu+3*sigma, 1000)
normpdf = lambda x, mu, sigma: np.exp(-(x-mu)**2/(2*sigma**2))/np.sqrt(2*np.pi)/sigma
plt.plot(xaxis, normpdf(xaxis, mu, sigma));
```
